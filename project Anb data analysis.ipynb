```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("compressed_data.csv")
```

```python
# check for duplicate values
data.duplicated().sum()
```

```python
data.drop_duplicates()
```

```python
data.drop(columns = ['house_rules', 'license'], axis = 1, inplace = True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Convert 'last review' to datetime and handle errors
data['last review'] = pd.to_datetime(data['last review'], errors='coerce')
# Fill missing values
data.fillna({'reviews per month': data['last review'].mean(), 'last review': data['last review'].mean(), 'review rate number': data['review rate number'].mean()}, inplace=True)
# Drop records with missing 'name' or 'host name'
data.dropna(subset=['NAME', 'host name'], inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.loc[data['availability 365'] < 0, 'availability 365'] = 0
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
print(data['minimum nights'].dtype)
print("*"*20)
print(data['minimum nights'].mode())
```

```python
data.fillna({'minimum nights': data['minimum nights'].mean()},inplace=True)
```

```python
data['minimum nights'].describe()
```

```python
# Remove dollar signs and convert to float
data['price'] = data['price'].replace('[\$,]', '', regex=True).astype(float)
data['service fee'] = data['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
data['price'].head()
```

# Visualization

# Plot the distribution of listing prices

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=50, kde=True, color='red') # Set histogram color
plt.title('Distribution of Listing Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

# Room Type Analysis
Analyze the distribution of different room types.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='room type', data=data , color='hotpink')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

# Neighborhood Analysis
Examine how listings are distributed across different neighborhoods

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='neighbourhood group', data=data,color="lightgreen",order=data['neighbourhood group'].value_counts().index, palette="viridis")
plt.title('Number of Listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

# Price vs. Room Type
Visualize the relationship between price and room type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', hue='room type', data=data, palette='Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

# Reviews Over Time
Plot the number of reviews over time.

```python
data['last review'] = pd.to_datetime(data['last review'])
reviews_over_time = data.groupby(data['last review'].dt.to_period('m')).size()
plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
